```python
import pandas as pd
import numpy as np
```

```python
url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8'

table, = pd.read_html(url, match='Коефіцієнт народжуваності в регіонах України')

table
```

```python
## Вивести перші рядки таблиці за допомогою методу head
table.head()
```

```python
## Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
table.shape
```

```python
## Замініть у таблиці значення "—" на значення NaN
table = table.replace('—', np.nan)
table
```

```python
## Визначте типи всіх стовпців за допомогою dataframe.dtypes
table.dtypes
```

```python
## Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
def to_numeric_or_leave_unchanged(x):
    try:
        return pd.to_numeric(x)
    except:
        return x

table = table.applymap(to_numeric_or_leave_unchanged)
table.dtypes
```

```python
## Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
for column in table.columns:
  null_cells = table[column].isnull().sum()
  total_cells = len(table[column])
  fillness = f'{round(null_cells / total_cells * 100, 2)}%'
  print(column, fillness)
```

```python
## Видаліть з таблиці дані по всій країні, останній рядок таблиці
table = table.drop(table.index[-1])

table
```

```python
## Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
table = table.fillna(table.mean(numeric_only=True)).round()

table
```

```python
## Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
table.loc[table['2019'] > table['2019'].mean(), 'Регіон']
```

```python
## У якому регіоні була найвища народжуваність у 2014 році?
table.loc[table['2014'].idxmax(), 'Регіон']
```

```python
## Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
import matplotlib.pyplot as plt


plt.barh(
    table['Регіон'],
    table['2019'],
    height=0.7,
)
plt.xlabel('Кількість народжень')
plt.ylabel('Регіон')
plt.title('Діаграма народжувальноті за 2019 рік')

plt.show()
```

# Візуалізація

```python
import seaborn as sns
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=table.drop(['Регіон'], axis=1).T)
plt.title('Population by region')
plt.legend(title='region', labels=table['Регіон'])
```

```python
year1 = '1990'
year2 = '2000'
plt.figure(figsize=(10, 8))
sns.scatterplot(x=year1, y=year2, data=table, hue='Регіон', palette='viridis')
plt.title(f'population Comparison between {year1} and {year2}')
plt.xlabel(f'population in {year1}')
plt.ylabel(f'population in {year2}')
```

```python
transp_table = table.set_index('Регіон').T

plt.figure(figsize=(15, 20))
ax = transp_table.plot(kind='bar', stacked=True, cmap='viridis')
plt.title('Population Distribution by Region Over Time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(title='Region', loc='upper right', bbox_to_anchor=(1.2, 1))
plt.xticks(rotation=45)
plt.show()
```
