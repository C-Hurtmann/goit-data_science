```python
import pandas as pd
import numpy as np
```

```python
## Прочитайте файл 2017_jun_final.csv за допомогою методу read_csv
table = pd.read_csv('2017_jun_final.csv')
## Прочитайте отриману таблицю, використовуючи метод head
table.head()
```

```python
## Визначте розмір таблиці за допомогою методу shape
table.shape
```

```python
## Визначте типи всіх стовпців за допомогою dataframe.dtypes
table.dtypes
```

```python
## Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
def show_fillness():
  for column in table.columns:
    null_cells = table[column].isnull().sum()
    total_cells = len(table[column])
    fillness = f'{round(null_cells / total_cells * 100, 2)}%'
    print(column, fillness)

show_fillness()
```

```python
## Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"
protected_column = table['Язык.программирования']
table = table.dropna(axis=1)
table = pd.concat([table, protected_column], axis=1)
## Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"
show_fillness()
```

```python
## Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
table = table.dropna(axis=0)

show_fillness()
```

```python
## Визначте новий розмір таблиці за допомогою методу shape
table.shape
```

```python
## Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
python_data = table.loc[table['Язык.программирования'] == 'Python']

python_data
```

```python
## Визначте розмір таблиці python_data за допомогою методу shape
python_data.shape
```

```python
## Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
## Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
salary_extremums = python_data.groupby('Должность')['Зарплата.в.месяц'].agg(['min', 'max'])

salary_extremums
```

```python
## Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"
avg_salary_per_position = python_data.groupby('Должность')['Зарплата.в.месяц'].agg([np.mean]).round()

def fill_avg_salary(row):
  current_position = row['Должность']
  return avg_salary_per_position.loc[current_position, 'mean']

python_data['avg.salary'] = python_data.apply(fill_avg_salary, axis=1).astype(int)

python_data
```

```python
## Створіть описову статистику за допомогою методу describe для нового стовпчика.

description = python_data['avg.salary'].describe()

description
```

```python
## Збережіть отриману таблицю в CSV файл
python_data.to_csv('2017_jun_final_cleaned.csv')
```

# Візуалізація

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

```python
fetch = table.loc[table['Должность'] == 'Junior Software Engineer'].groupby('Язык.программирования')['Зарплата.в.месяц'].agg(['min', 'max']).reset_index()
melted_fetch = fetch.melt(id_vars='Язык.программирования', var_name='extremum', value_name='salary')
colors = ["#3498db", "#e74c3c"]
sns.barplot(x='salary', y='Язык.программирования', hue='extremum', data=melted_fetch, palette=colors)
plt.ylabel('language')
plt.title('Extremums of developers salaries')
```

```python
fetch = python_data.groupby('Должность')['N'].agg([('qty', 'count')]).reset_index()
explode = [0, 0, 0, 0.15, 0]
plt.pie(fetch['qty'], labels=fetch['Должность'], explode=explode, autopct='%1.1f%%', startangle=100, )
plt.title('Python developers diversity')
plt.show()
```

```python
fetch = table.loc[table['Должность'] == 'Junior Software Engineer']
sns.violinplot(x='Пол', y='salary', data=fetch)
plt.title('Salary distribution by gender for junior software engineers')
```
